```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import time
```

```python
num_data=1000

x=init.uniform_(torch.Tensor(num_data, 1), -5, 5)

noise = init.normal_(torch.FloatTensor(num_data,1),std=2)

y1=x**4+x**2
y1_noise=(x+noise)**4+(x+noise)**2

y2=x**2+x
y2_noise=(x+noise)**2+(x+noise)
```

```python
model = nn.Linear(in_features=1, out_features=1) 

loss_func=nn.MSELoss()

learning_rate=0.01

optimizer=optim.SGD(model.parameters(),lr=learning_rate )
```

```python
label1=y1_noise

num_epoch=500
losses=[]

start=time.time()

if torch.cuda.is_available():
    model=model.cuda()
    x=x.cuda()
    label=y1_noise.cuda()
    
for i in range(num_epoch):
    
    optimizer.zero_grad()
    
    output=model(x)
    
    loss=loss_func(output, label1)
    loss.backward()
    
    optimizer.step()
    
    if i % 10 == 0:
        print(loss.data.cpu()) #Tensor data의 값만 출력하고 싶을 경우 .data 사용
    losses.append(loss.data.cpu())

print(time.time()-start)
```

```python
label2=y2_noise

num_epoch=500
losses=[]

start=time.time()

if torch.cuda.is_available():
    model=model.cuda()
    x=x.cuda()
    label=y2_noise.cuda()
    
for i in range(num_epoch):
    
    optimizer.zero_grad()
    
    output=model(x)
    
    loss=loss_func(output, label2)
    loss.backward()
    
    optimizer.step()
    
    if i % 10 == 0:
        print(loss.data.cpu()) #Tensor data의 값만 출력하고 싶을 경우 .data 사용
    losses.append(loss.data.cpu())

print(time.time()-start)
#y2가 y1보다 빠름
```
